# Data Freshness Check

Query lakehouse tables to verify latest data dates and row counts.

- Input: Lakehouse delta tables
- Output: Print freshness summary

```python
from pyspark.sql import functions as F

# Read from lakehouse table using three-part naming
df = spark.sql("""
    SELECT 
        max(date_key) as latest_date,
        min(date_key) as earliest_date,
        count(*) as total_rows
    FROM LakehouseName.schema.table_name
""")

# Print results (display() does not show output in fab job run)
pandas_df = df.toPandas()
print(pandas_df)
print(f"\nLatest date: {pandas_df.iloc[0]['latest_date']}")
```

## Write results to lakehouse

```python
# Write DataFrame to lakehouse table
# Three-part naming: LakehouseName.schema.table
df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("LakehouseName.schema.freshness_log")
print("Written to lakehouse")
```
